```python
import pandas as pd
import numpy as np
import random
import string
import time
```

```python
countries = ['\'Russia\'', '\'Japan\'', '\'Brazilia\'', '\'Ukrain\'', '\'USA\'', '\'Spain\'', '\'France\'', '\'China\'']
cities = ['\'Moscow\'', '\'Tokyo\'', '\'Brazilia\'', '\'Kiev\'', '\'Washington\'', '\'Madrid\'', '\'Paris\'', '\'Beijing\'']
sports = ['\'swimming\'', '\'tennis\'', '\'judo\'', '\'sambo\'', '\'figure skating\'', '\'skying\'', '\'gymnastic\'']
```

```python
def generate_name():
    length1 = random.randint(2, 10)
    letters = string.ascii_lowercase
    res1 = ''.join(random.choice(letters) for i in range(length1))
    length2 = random.randint(2, 10)
    res2 = ''.join(random.choice(letters) for i in range(length2))
    res = (res1 + ' ' + res2).title()
    return '\'' + res + '\''
```

```python
generate_name()
```

```python
def generate_date():
    y = random.randint(1960, 2005)
    m = random.randint(1, 13)
    d = random.randint(1, 28)
    return (str)(d) + '.' + (str)(m) + '.' + (str)(y)
```

```python
def fill_table(df, name_table):
    res = []
    for i in range(df.shape[0]):
        s = 'insert into ' + name_table + ' values (' + ', '.join(df.iloc[i]) + ');'
        res.append(s)
    return res
```

```python
fill_list = []
```

```python
fill_list = fill_list + fill_table(countries_df, 'countries')
fill_list
```

```python
fill_list = fill_list + fill_table(olympics_df, 'olympics')
fill_list
```

```python
fill_list = fill_list + fill_table(events_df, 'events')
fill_list
```

```python
fill_list = fill_list + fill_table(players_df, 'players')
fill_list 
```

```python
fill_list = fill_list + fill_table(results_df, 'results')
fill_list
```
